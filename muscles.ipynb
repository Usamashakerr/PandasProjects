# 1. Data Cleaning #

```python
import pandas as pd
```

```python
df=pd.read_csv('D:/Datasets/Protien intake and muscles.csv')
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.fillna(df.mean(numeric_only=True),inplace=True)
```

```python
df.isnull().sum()
```

```python
df[['Author', 'Year']]=df['Author and Year'].str.extract(r'^(.*?)\s\((\d{4})\)')
```

```python
df
```

```python
df=df.drop(columns='Author and Year')
```

```python
df
```

```python
df['Energy balance']=df['Energy balance'].replace({'Intake > Expenditure' : 'Intake more than Expenditure', 'Intake < Expenditure': 'Intake less than Expenditure','Intake = Expenditure':'Intake equal Expenditure'})
```
